# David Hume Word Clouds

```python
%matplotlib inline
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

## Masks

Load the grayscale images and convert them to pure black and white:

```python
hume1 = Image.open('images/hume1.png')
hume1 = np.array(hume1)
hume1 = np.array([[255 if y>220 else 0 for y in x] for x in hume1])

plt.imshow(hume1, cmap='Greys');
```

```python
hume2 = Image.open('images/hume2.png')
hume2 = np.array(hume2)
hume2 = np.array([[255 if y[0]>230 else 0 for y in x] for x in hume2])

plt.imshow(hume2, cmap='Greys');
```

## Configure Word Clouds

Load the stop words and initialize new WordCloud objects:

```python
with open('texts/stopwords.txt', 'rt') as fh:
    stopwords = set(fh.read().split('\n'))
```

```python
background = 'black'
wc1 = WordCloud(width = hume1.shape[1], height = hume1.shape[0],
                stopwords=stopwords, mask=hume1, background_color=background)
wc2 = WordCloud(width = hume2.shape[1], height = hume2.shape[0],
                stopwords=stopwords, mask=hume2, background_color=background)
```

Setup plot configuration:

```python
def plot(wordcloud, text, name=None, scale=50):
    image = wordcloud.generate(text)
    size = np.array(wordcloud.mask.shape)/scale
    plt.figure(figsize=size)
    plt.imshow(image, interpolation='bilinear')
    plt.tight_layout(pad=0)
    plt.axis('off')
    
    if name is not None:
        plt.savefig(name, facecolor=background, bbox_inches='tight')
        
    plt.show()
```

## Enquiry into Human Understanding

```python
with open('texts/EHU.txt', 'rt') as fh:
    s = fh.read()
    
plot(wc1, s, 'wordclouds/ehu1.png', scale=70), plot(wc2, s, 'wordclouds/ehu2.png');
```

## Enquiry into the Principles of Morals

```python
with open('texts/EPM.txt', 'rt') as fh:
    s = fh.read()

plot(wc1, s, 'wordclouds/epm1.png', scale=70), plot(wc2, s, 'wordclouds/epm2.png');
```
